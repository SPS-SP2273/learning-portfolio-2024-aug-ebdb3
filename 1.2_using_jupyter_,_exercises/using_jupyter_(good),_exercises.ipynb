```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

1. first cell: squares = a list of $i^2$ for $1 \leq i \leq 10$. second cell = squares is a list. for $1 \leq i \leq 10$, add $i^2$ to the list 'squares'.
2. second cell has a higher mean and standard deviation - takes more time on average and larger range of results for time taken to produce output because it adds $i^2$ to the list individually for every i value. the first cell adds them together, taking a lesser time.
