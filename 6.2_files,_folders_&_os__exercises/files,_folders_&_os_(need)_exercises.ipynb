```python
import os
import glob
import shutil
```

### Task 1

```python
date_conversion = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
```

```python
filename = "oct-08_singapore.txt"
filename_split = filename.split("-")

for string in filename_split:
    filename_converted = date_conversion[filename_split[0]]
        # run the first item in filename_split (oct) thru date_conversion dictionary to get 10

filename_split2 = filename_split[1].split("_")

print(filename_converted + "-" + filename_split2[0] + "-" + filename_split2[1])
```

### Task 2

```python
def rename_my_file(old_file_name):
    OFN_split = old_file_name.split("-")
    for string in OFN_split:
        OFN_converted = date_conversion[OFN_split[0]]
    OFN_split2 = OFN_split[1].split("_")
    return(OFN_converted + "-" + OFN_split2[0] + "-" + OFN_split2[1])

print(rename_my_file("oct-08_singapore.txt"))
```

```python
for task6_sourcepath in glob.glob('os-collaboration-exercise-data\\*Cairo.*'):
    # listing all files in os-collab(etc) with Cairo in title 
    # "." after Cairo needed to ensure only files are listed
    try:
        task6_destination = os.path.join('.', 'os-collaboration-exercise-data', 'Cairo')
        shutil.copy(task6_sourcepath, task6_destination)
        print(f'Copied {task6_sourcepath} to {task6_destination}')
    except FileExistsError:
        print(f'{task6_destination} already exists')
```

### Task 7

```python
city_list = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in city_list:
    text = os.path.join('.', 'os-collaboration-exercise-data', "*" + city + ".*")
        # use os.path.join and not a string otherwise it doesnt work TT

    for task7_sourcepath in glob.glob(text, recursive = True):
        try:
            filename = task7_sourcepath.split("\\")[-1]
            task7_destination = os.path.join('.', 'os-collaboration-exercise-data', city, filename) 
                # shutil.move doesnt move files if they're alr in the destination file - putting full path (incl. filename) overrides this
            print(task7_destination)
            shutil.move(task7_sourcepath, task7_destination)
            print(f'Moved {filename} to {task7_destination}')
        except FileExistsError:
            print(f'{filename} already exists in {task7_destination}')
```
